```python
import PyPDF2
import re

def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num]
            text += page.extract_text()
    return text

pdf_path = '2023-2024-spring-courses.pdf'
text = extract_text_from_pdf(pdf_path)
```

```python
lines = text.split('\n')
print(len(lines))
```

```python
classes = []
for l in lines:
    if ' — ' in l:
        classes.append(l)
        
len(classes)
```

```python
parsed_classes = []
for c in classes:
    pos = c.find('—')
    t = c[0:pos-1]
    parsed_classes.append(t)
    print(t)
```

```python
anomaly = []

for p in parsed_classes:
    if 'Last Taught:' in p:
        anomaly.append(p)
        print(p)
    elif 'Repeatable' in p:
        anomaly.append(p)
```

```python
cross_listed = []
for p in parsed_classes:
    if '/' in p:
        cross_listed.append(p)
        print(p)
```

```python
classes_list = []
for c in cross_listed:
    divide(c, classes = classes_list)
```

```python
master_classes = []
for p in parsed_classes:
    if (p not in anomaly) & (p not in cross_listed):
        master_classes.append(p)
        print(p)
```

```python
def divide(text, classes):
    slash = None
    if '/' in text:
        slash = text.find('/')
    else:
        classes.append(text)
        return
        
    number = text[-3:]
    class1 = text[0:slash] + " " + number
    classes.append(class1)
    class2 = text[slash+1:len(text)-4] + " " + number
    if '/' in class2:
        divide(class2, classes_list)
    else:
        classes.append(class2) 
```
